```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('housing.csv')
```

```python
df=df.dropna()
df = df.reset_index()
```

I dropped all rows with a missing value in any column.

```python
one_hot = pd.get_dummies(df['ocean_proximity'])
new_df=df.drop('ocean_proximity',axis=1)
correlation=new_df.join(one_hot)
```

Predictor variable is the median house value for properties within the block of the house (median_house_value).

```python
df = pd.read_csv('housing.csv')
df = df.dropna()
one_hot = pd.get_dummies(df['ocean_proximity'])
x=new_df.drop(columns=['median_house_value']).join(one_hot)
y = new_df['median_house_value']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3,random_state = 99)

mindegree = 0
maxdegree = 5
train_error = []
test_error = []
for deg in range(mindegree,maxdegree):
    model = make_pipeline(PolynomialFeatures(deg),LinearRegression())
    model.fit(x_train,y_train)
    train_error.append(mean_squared_error(y_train,model.predict(x_train)))
    test_error.append(mean_squared_error(y_test,model.predict(x_test)))
print(train_error,test_error)
plt.plot(np.arange(mindegree,maxdegree), train_error, color='green', label='train')
plt.plot(np.arange(mindegree,maxdegree), test_error, color='red', label='test')
plt.ylabel('mean squared error')
plt.xlabel('degree')
plt.legend(loc='upper left')
```

Optimal Degree for Polynomial Regression: 3

```python
df = pd.read_csv('housing.csv')
df = df.dropna()
df = df.reset_index()
one_hot = pd.get_dummies(df['ocean_proximity'])
Y = df['median_house_value']
numerical = df.drop(columns = ['ocean_proximity', 'median_house_value'])
model = make_pipeline(PolynomialFeatures(3), StandardScaler())
model.fit(numerical)
X = one_hot.join(pd.DataFrame(model.transform(numerical)))
model = RidgeCV(cv = 10, alphas = [0.0001,0.001,0.01,0.1,1,10])
model.fit(X,Y)
print('R-Squared:', model.score(X, Y)) 
print('Alpha:', model.steps[1][1].alpha_)
```
